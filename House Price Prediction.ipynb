```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data= pd.read_csv('ParisHousing.csv')
data
```

# Exploratory data analysis

```python
data.describe()
```

```python
#Checking their is any null value or not
data.isnull().sum()
```

```python
data.info()
```

```python
data.corr()
```

```python
# To find out their is any outliers in the attribute
plt.figure(figsize=(10,4))
sns.boxplot(data.iloc[:,:-1])
plt.xticks(rotation=45)
plt.show()
```

# Train- test split

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_trans= scalar.fit_transform(x_train)
print(x_trans.shape)
print(y_train.shape)
```

# Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_trans, y_train)
```

```python
#Creating a pipeline for first model
from sklearn.pipeline import make_pipeline
pipe1= make_pipeline(scalar, model1)
```

```python
pipe1.fit(x_train, y_train)
```

```python
pred1 = pipe1.predict(x_test)
```

```python
pred1
```

### # Cheking the accuracy of Multiple Regression model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae1 = mean_absolute_error(y_test, pred1)
msa1 = mean_squared_error(y_test, pred1)
```

```python
print(mae1)
print(msa1)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, pred1))
```

```python
residual = y_test- pred1
```

```python
#Residual plot
#Residual plot follows normal distribution i.e. model is perfectly fitted  
sns.displot(residual)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
model2 = Ridge()
model2.fit(x_trans, y_train)
```

```python
#Creating a pipeline for second model
from sklearn.pipeline import make_pipeline
pipe2 = make_pipeline(scalar, model2)
```

```python
pipe2.fit(x_train, y_train)
```

```python
pred2= pipe2.predict(x_test)
```

### # Cheking the accuracy of Ridge Regression model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae1 = mean_absolute_error(y_test, pred2)
msa1 = mean_squared_error(y_test, pred2)
```

```python
print(mae1)
print(msa1)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, pred2))
```

```python
residual = y_test- pred2
```

```python
# Residual plot
#Residual plot follows normal distribution i.e. model is perfectly fitted 
sns.displot(residual)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
model3 = Lasso()
model3.fit(x_train, y_train)
```

```python
#Creating a pipeline for first model
from sklearn.pipeline import make_pipeline
pipe3= make_pipeline(scalar, model3)
```

```python
pipe3.fit(x_train, y_train)
```

```python
pred3= pipe3.predict(x_test)
```

```python
residual= y_test- pred3
```

```python
#Residual plot
#Residual plot follows normal distribution i.e. model is perfectly fitted 
sns.displot(residual)
```

### # Cheking the accuracy of Lasso Regression model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae3 = mean_absolute_error(y_test, pred3)
msa3 = mean_squared_error(y_test, pred3)
```

```python
print(mae3)
print(msa3)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, pred3))
```

```python
# All three model are fitted good. but Lasso regression gives more accuracy. 
print("No Regularization:", r2_score(y_test, pred1))
print("Ridge:", r2_score(y_test, pred2))
print("Lasso:", r2_score(y_test, pred3))
```
